```python
import pandas as pd
import glob

# put the excel files in the same file
path = "."
file_list = glob.glob(path + "/*.xlsx")
print(file_list)
excl_list = []
 
for file in file_list:
    excl_list.append(pd.read_excel(file,dtype='str'))
    
excl_merged = pd.DataFrame()

for excl_file in excl_list:
    excl_merged = excl_merged.append(
      excl_file, ignore_index=True)

excl_merged.to_excel('ServiceNowQuestions.xlsx', index=False,encoding='utf-8')
```

```python
excl_merged
```

```python
#removing the unvalid values
excl_merged_without_nan = excl_merged.dropna() 
excl_merged_without_nan
```

```python
excl_merged_without_nan = excl_merged_without_nan.reset_index()
```

```python
excl_merged_without_nan
```

```python
#looping over Tag column in the index 1
for column in excl_merged_without_nan.columns[1:2]:
    for index in range(len(excl_merged_without_nan[column])):
        #slicing the column
        tag_to_slice = excl_merged_without_nan.at[index,'Tags']
        excl_merged_without_nan.at[index,'Tags'] = tag_to_slice[:tag_to_slice.find(',')]
```

```python
excl_merged_without_nan
```

```python
unique = excl_merged_without_nan['Tags'].unique()
```

```python
for column in excl_merged_without_nan.columns[:1]:
    for index in range(len(excl_merged_without_nan[column])):
        #slicing the column
        excl_merged_without_nan.at[index,'index'] = index
```

```python
#convert html Questions to normal
for column in excl_merged_without_nan.columns[3:4]:
    for index in range(len(excl_merged_without_nan[column])):
        #slicing the column
        soup = BeautifulSoup(""+excl_merged_without_nan.at[index,'Question']+"")
        soup_ = soup.get_text()
        soup_ = soup_.replace("\n", "")
        soup_ = soup_.replace("\xa0", "")
        excl_merged_without_nan.at[index,'Question'] = soup_
```

```python
#convert html Answers to normal
for column in excl_merged_without_nan.columns[4:]:
    for index in range(len(excl_merged_without_nan[column])):
        #slicing the column
        soup = BeautifulSoup(""+excl_merged_without_nan.at[index,'Answer']+"")
        soup_ = soup.get_text()
        soup_ = soup_.replace("\n", "")
        soup_ = soup_.replace("\xa0", "")
        excl_merged_without_nan.at[index,'Answer'] = soup_
```
